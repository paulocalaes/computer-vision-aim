```python
# Importação das bibliotecas
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('lenna.jpeg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
print(image.shape)
```

```python
img = cv2.imread('lenna.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


scale_percent = 50 # percent of original size try 25, 10, 5
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
plt.imshow(resized, cmap='gray')
print(resized.shape)
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('lenna.jpeg',0)
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
# import Opencv 
import cv2 
import numpy as np 
  
# read a image using imread 
img = cv2.imread('lenna.jpeg', 0) 
  
# creating a Histograms Equalization 
# of a image using cv2.equalizeHist() 
equ = cv2.equalizeHist(img) 
  
# stacking images side-by-side 
res = np.hstack((img, equ)) 
  
# show image input vs output 
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
print(image.shape)
```

```python
import cv2 as cv
import numpy as np
img = cv.imread('lenna.jpeg')
norm_img = np.zeros((800,800))
final_img = cv.normalize(img,  norm_img, 0, 255, cv.NORM_MINMAX)
plt.imshow(cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB))
```
