```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imgGray = cv2.imread('xray2.png')
imgGray = cv2.cvtColor(imgGray, cv2.COLOR_BGR2GRAY)

template = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])

dst = cv2.filter2D(imgGray,-1,template)

newImg =  np.zeros((208,157))

newImg = dst > 160

# stacking images side-by-side 
res = np.hstack((imgGray, dst, newImg)) 
plt.imshow(res, cmap='gray')
print(res.shape)
```
