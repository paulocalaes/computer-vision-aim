```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imgGray = cv2.imread('lines.png')
imgGray = cv2.cvtColor(imgGray, cv2.COLOR_BGR2GRAY)

template = np.array([[-1,-1,-1], [2,2,2], [-1,-1,-1]]) #horizontal
#template = np.array([[-1,2,-1], [-1,2,-1], [-1,2,-1]]) #vertical
#template = np.array([[2,-1,-1], [-1,2,-1], [-1,-1,2]]) # +45
#template = np.array([[-1,-1,2], [-1,2,-1], [2,-1,-1]]) # -45


kernel = np.ones((6,6),np.uint8)
erosion = cv2.erode(imgGray,kernel,iterations = 1)

horizontal = cv2.filter2D(erosion,-1,template)

# stacking images side-by-side 
res = np.hstack((imgGray, horizontal)) 
plt.imshow(res, cmap='gray')
print(imgGray.shape)
```

```python
imgGray = cv2.imread('lines.png')
imgGray = cv2.cvtColor(imgGray, cv2.COLOR_BGR2GRAY)

templateH = np.array([[-1,-1,-1], [2,2,2], [-1,-1,-1]]) #horizontal
templateV = np.array([[-1,2,-1], [-1,2,-1], [-1,2,-1]]) #vertical
template45 = np.array([[2,-1,-1], [-1,2,-1], [-1,-1,2]]) # +45
template452 = np.array([[-1,-1,2], [-1,2,-1], [2,-1,-1]]) # -45


kernel = np.ones((6,6),np.uint8)
erosion = cv2.erode(imgGray,kernel,iterations = 1)

horizontal = cv2.filter2D(erosion,-1,templateH)
vertical = cv2.filter2D(erosion,-1,templateV)

kernel = np.ones((4,4),np.uint8)
erosion = cv2.erode(imgGray,kernel,iterations = 1)

inc1 = cv2.filter2D(erosion,-1,template45)

kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(imgGray,kernel,iterations = 1)
inc2 = cv2.filter2D(erosion,-1,template452)

# stacking images side-by-side 
res = np.hstack(( horizontal,vertical,inc1,inc2)) 
plt.imshow(res, cmap='gray')
print(imgGray.shape)
```
