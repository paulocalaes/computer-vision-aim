# Filtros

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Filtro Passa Alta

```python
imgGray = cv2.imread('lenna.jpeg')
imgGray = cv2.cvtColor(imgGray, cv2.COLOR_BGR2GRAY)

template = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
template = np.array([[0,-1,0], [-1,4,-1], [0,-1,0]])
dst = cv2.filter2D(imgGray,-1,template)

plt.subplot(121),plt.imshow(imgGray, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst, cmap='gray'),plt.title('Passa Alta')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
dst = cv2.filter2D(imgGray,-1,template)
newImage = 255 - dst


plt.subplot(121),plt.imshow(imgGray, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(newImage, cmap='gray'),plt.title('255 - Passa Alta')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Filtro da Mediana

```python

img = cv2.imread('lenna.jpeg', 1)
row,col,ch = img.shape
p = 0.5
a = 0.009
noisy = img

  # Salt mode
num_salt = np.ceil(a * img.size * p)
coords = [np.random.randint(0, i - 1, int(num_salt))
          for i in img.shape]
noisy[tuple(coords)] = 1

  # Pepper mode
num_pepper = np.ceil(a * img.size * (1. - p))
coords = [np.random.randint(0, i - 1, int(num_pepper))
          for i in img.shape]
noisy[tuple(coords)] = 0

median_blur= cv2.medianBlur(noisy, 3)

plt.subplot(121),plt.imshow(cv2.cvtColor(noisy, cv2.COLOR_BGR2RGB)),plt.title('Noisy')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)),plt.title('255 - Passa Alta')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Filtro da Gaussiana

```python
blur = cv2.GaussianBlur(noisy,(5,5),3)

plt.subplot(121),plt.imshow(cv2.cvtColor(noisy, cv2.COLOR_BGR2RGB)),plt.title('Noisy')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)),plt.title('255 - Passa Alta')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Filtro Bilateral

```python
blur = cv2.bilateralFilter(img,9,75,75)

plt.subplot(121),plt.imshow(cv2.cvtColor(noisy, cv2.COLOR_BGR2RGB)),plt.title('Noisy')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)),plt.title('255 - Passa Alta')
plt.xticks([]), plt.yticks([])
plt.show()
```
